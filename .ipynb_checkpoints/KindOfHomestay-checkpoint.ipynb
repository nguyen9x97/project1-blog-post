# Question 1: What kinds of homestay are there?
This project is using Seattle Airbnb dataset: https://www.kaggle.com/datasets/airbnb/seattle

Firstly, let's look at the dataset

```python
# Import some neccessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
%matplotlib inline
```

```python
df = pd.read_csv('listings.csv')
df.head()
```

Now to look at the first question: What kinds of homestay are there? Look at the `property_type` field, you can see there are a lot of homestay types. Below you can take a look at each type.

```python
property_type_vals = df['property_type'].value_counts().reset_index()
property_type_vals.rename(columns={'index': 'Homestay type', 'property_type': 'count'}, inplace=True)
property_type_vals
```

```python
# We might also look at the percentage
property_type_vals['perc'] = property_type_vals['count'] / np.sum(property_type_vals['count'])
property_type_vals
```

Let's visualize it in a chart

```python
property_type_vals.plot(kind='bar', x='Homestay type', y='perc')
```

As you can see, there are 16 types of homestay in total in this dataset. However, the proportion of House and Apartment accounts for the most.

Therefore, I can conclude that **House** and **Apartment** are the most common
